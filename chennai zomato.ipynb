# 2 we r going to check Address

```python
raw_dataset["Address"].value_counts()# here we hv delivery only as the name and location of the address which is useless so 
#we remove that.
```

```python
#1. we strip extra spaces from all values in our address column and always check if u want to modify any column.
raw_dataset["Address"].str.strip()
```

```python
# 2 . no we r going to create a condition to check if None and Delivery Only exists and store them as useless data
raw_dataset["Address"].str.strip()
useless_data=raw_dataset["Address"].isin(["None","Delivery Only"])#
# we r using .isin() method to check for 2 unwanted strings at a time instead of ==
```

```python
useless_data# this gives a boolean series
```

```python
raw_dataset=raw_dataset[~useless_data]
```

```python
raw_dataset
```

# 3 we r going to drop the location column as it is already present in Address


```python
raw_dataset.drop("Location",axis=1,inplace=True)
```

# 4 we r going to modify cuisine by removing special characters
